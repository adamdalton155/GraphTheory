<h2>Topic Two - Specification of the Graph Isomorphism Problem</h2>
<hr>

The Graph Isomorphism problem is a computational problem which attempts to determine whether two finite graphs are isomorphic or not. Isomorphic meaning that two graphs contain the same number of vertices which are also connected in the same way. For two graphs G and H to be proven to be isomorphic, you must give the bijection f and check the condition on numbers of edges (and loops) for all pairs of vertices v, w ∈ V (G)

The Graph Isomorphism problem is currently not known to solvable. The reason for this is the problem cannot be solved in polynominal time or to be NP-Complete. What this means the execution time is either given by a polynomial on the size of the input or can be bounded by a polynomial, and NP-Complete means the complexity class of decision problems for which answers can be checked for correctness, given a cetificate by an algorithm whose run time is polynomial in the size of the input (that is, it is NP) and no other NP problem is more than a polynomial factor harder.

Due to the fact that this problem isnt known to be NP-Complete, researches have defined a new class GI, a set of problems with polynominal-time Turing reduction. If the graph is solvable in polynominal-time, GI would be equal to P whereas if the problem is NP-complete, GI would be equal to NP meaning all the problems in NP are solved in 
quasi-polynomial time. If GI is equal to P, the problem is then known as GI-Hard.

In 2015, a theoretical computer scientist of the university of the university of Chicago named László Babai claimed to have solved Graph Isomorphism problem. His new algorithm at the time was more efficient the previous algorithm which held the record for over thirty years. He claimed his algorithm was able to solve the GI problem in quasi-polynominal time. Then on January 4th, László Babai shocked the Computer Science community by retracting the claim he made back in 2015 for the scientist to announce on the 9th that he in fact solved the flaw in his proof. Over the years there have been multiple algorithms presented that claimed the be the one to fully solve the GI problem but it still has not officially been solved

<h2>Topic Four - Explanation of the main reason computer scientists are interested in the complexity of the Graph Isomorphism Problem</h2>
<hr>

The main reason that computer scientists are so interested in the complexity of the Graph Isomorphism Problem is that after all these decades, we still do not know whether the problem can even be solved. Thousdands of other computational problems have been given the label hard or easy to solve but GI seems to be a middle state of both and is one of the few problems that isn't NP complete. 

In 2015 when László Babai announced his quasipolynomial time algorithm, that was the first major breakthrough in around 32 years. The previous accepted theoretical algorithm was Babai and another computer scientist Eugene M. Luks in 1983. This algorithm was based Luks previous work.

There are other competing algorithms for Graph Isomorphism such as McKay, Brendan D. (1981), "Practical graph isomorphism", Schmidt, Douglas C.; Druffel, Larry E. (1976), "A fast backtracking algorithm to test directed graphs for isomorphism using distance matrices" and Ullman, Julian R. (1976), "An algorithm for subgraph isomorphism". The drawback to these is their exponential time performance in the worst case.


Graph Isomorphism has been such a long standing problem. It takes years for any major breakthroughs to happen and while they do occur, they still haven't proved anything. Thats why so many computer scientists are interested in the Graph Isomorphism problem. Its not hard or easy, and so many have made attempts but nothing have fully came from it.
