```python
#Bubble in sort in action woooooo as a function

def bubble_Sort(L):

    #Keep track of number of comparisons
    num_comparisons = 0

    #Bubble every biggest element up
    for j in range (len(L)-1):
        #Keep track of any swaps
        swapped = False
        #Compare all elements that are side by side
        for i in range(len(L) -1):
            #Compare in ith element with the (i+1)th.
            if L[i] > L[i+1]:
                #Swap the elements
                L[i], L[i+1] = L[i+1], L[i]
                #Keep track of the swap
                swapped = True
            #Add a comparison
            num_comparisons = num_comparisons + 1

        #Quit if we didnt make any swaps
        if not swapped:
            break

    return num_comparisons
```
