```python
def siftDown(a, start, end):
    root = start
    
    while (2 * root + 1) <= end:
        child = (2 * root + 1)
        swap = root
        
        if a[swap] < a[child]:
            swap = child
            
        if child + 1 <= end and a[swap] < a[child + 1]:
            swap = child + 1
        
        if swap == root:
            return
        
        else:
            a[root], a[swap] = a[swap], a[root]
            root = swap
```

```python
def heapsort(a):
    count = len(a)
    
  
    end = count - 1
    
    while end > 0:
        a[0], a[end] = a[end], a[0]
        end = end - 1
        siftDown(a, 0, end)
```
