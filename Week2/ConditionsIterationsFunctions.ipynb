<h3>Functions</h3>

```python
def square(i):
    return i * i
```

```python
square(8)
```

```python
def power(i, j=2):
    result = 1
    for k in range(j):
        result = result * i
    return result
```

```python
power(9)
```

```python
power(9, 3)
```
