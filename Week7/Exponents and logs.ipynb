<h1>Big O</h1>
<hr>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.1, 10.1, 0.1)
y1 = x**2

y2 = x * np.log(x)
```

```python
plt.plot(x, y1, label='$^2$')
plt.plot(x, y2, label='$x \log(x)$')

plt.legend
```


$f(n) = n^2 $ is O(g(n))$ where $g(n) = n^3 

because

$ 1 \times g(n) > f(n) $ for all $n >= 1$

```python
n = np.arange(0.1, 10.1, 0.1)
plt.plot(n, n**2, label='$n^2$')
plt.plot(n, n**3, label='$n^3$')

plt.legend
```

Technically, $2n^2$ is O(n^2)$, even though....

```python
n = np.arange(0.1, 10.1, 0.1)
plt.plot(n, 2.0 * (n**2), label='$n^2$')
plt.plot(n, n**2, label='$n^3$')

plt.legend
```

$ n \times \log(n) $ + $n \times \log(n) = 2 n \log(n)$

$2 n \log(n)$ is $O(n \log(n))$
