```python
def siftDown(a, start, end):
    root = start
    
    while (2 * root + 1) <= end:
        child = (2 * root + 1)
        swap = root
        
        if a[swap] < a[child]:
            swap = child
            
        if child + 1 <= end and a[swap] < a[child + 1]:
            swap = child + 1
        
        if swap == root:
            return
        
        else:
            a[root], a[swap] = a[swap], a[root]
            root = swap
```
